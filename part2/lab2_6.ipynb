```python
import cv2
```

```python
import numpy as np
from matplotlib import pyplot
```

```python
img = cv2.imread("../photos/lab5_2.png")
```

```python
# make comparison of methods Harrisa, SIFT, SURF, FAST i ORB on 5_2 image
# Harris
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = np.float32(gray)
dst = cv2.cornerHarris(gray, 5, 5, 0.01)
dst = cv2.dilate(dst, None)
img_harris = img.copy() 
img_harris[dst>0.01*dst.max()]=[0,0,255]

corners = np.where(dst > 0.01 * dst.max())
for y, x in zip(*corners):
    cv2.line(img_harris, (x - 5, y), (x + 5, y), (0, 250, 0), 2)
    cv2.line(img_harris, (x, y - 5), (x, y + 5), (0, 250, 0), 2)

pyplot.figure(figsize=(10, 10))
pyplot.imshow(cv2.cvtColor(img_harris, cv2.COLOR_BGR2RGB))
pyplot.show()
    
```

```python
# SIFT
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
sift = cv2.SIFT_create()
kp = sift.detect(gray, None)
    
img2 = cv2.drawKeypoints(gray,kp,img,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
pyplot.figure(figsize=(10, 10))
pyplot.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
pyplot.show()

```

```python

# FAST
fast = cv2.FastFeatureDetector_create()
kp = fast.detect(gray, None)

img2 = cv2.drawKeypoints(gray, kp, img, color=(0,250,250))
pyplot.figure(figsize=(10, 10))
pyplot.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
pyplot.show()
```

```python

# ORB
orb = cv2.ORB_create()
kp = orb.detect(gray, None)
kp, des = orb.compute(img, kp)
img2 = cv2.drawKeypoints(gray, kp, img, color=(0,250,250))

pyplot.figure(figsize=(10, 10))
pyplot.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
pyplot.show()
```
